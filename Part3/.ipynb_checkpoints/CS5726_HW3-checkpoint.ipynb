### Question 1

```python
# Part f
seed(0)
lambda_star = 2
n = 20
B = 1000
reject = 0
for _ in range(B):
    samples = poisson(lam=lambda_star,size=n)
    mean = scipy.mean(samples)
    se = np.sqrt(mean/n)
    if lambda_star < mean-1.96*se or lambda_star > mean+1.96*se:
        reject+=1
print(reject)
```

```python
# Part h
seed(0)
lambda_star = 2
tau_star = log(lambda_star)
n = 20
B = 1000
reject = 0
for _ in range(B):
    samples = poisson(lam=lambda_star,size=n)
    mean = scipy.mean(samples)
    se = np.sqrt(1/(n*mean))
    if tau_star < log(mean)-1.96*se or tau_star > log(mean)+1.96*se:
        reject+=1
print(reject)
```

```python
# Part h
seed(0)
lambda_star = 2
tau_star = log(lambda_star)
n = 20
B = 1000
ci0 = []
ci1 = []
for _ in range(B):
    samples = poisson(lam=lambda_star,size=n)
    mean = scipy.mean(samples)
    se0 = np.sqrt(mean/n)
    se1 = np.sqrt(1/(n*mean))
    ci0.append(log(mean+1.96*se0)-log(mean-1.96*se0))
    ci1.append((log(mean)+1.96*se1)-(log(mean)-1.96*se1))
print(np.mean(ci0), np.mean(ci1))
```

```python
dfr = pd.read_csv('winequality-red.csv',sep=';')
dfw = pd.read_csv('winequality-white.csv',sep=';')
red = np.hstack((np.ones(len(dfr)), np.zeros(len(dfw))))
```

```python
df = pd.concat([dfr, dfw])
df['red'] = red
df.columns = df.columns.str.replace(' ', '_')
```

```python
df_train, df_test = train_test_split(df, train_size=5000, random_state=0)
df_train = df_train.reset_index(drop=True)
response = 'quality'

print(df_train.shape)
print(df_test.shape)
print(df.shape)
```

```python
def fitmodel(S,data,response):
    return smf.ols(response+' ~ '+('+'.join(S) if len(S)>0 else '1'), data=data).fit()
def CV_MSE(response, features, data, folds=20):
    idxs = np.arange(len(data))/int(np.ceil(len(data)/float(folds)))
    idxs = idxs.astype(int)
    
    fs = max(idxs)+1
    mses = np.zeros(fs)
    for f in range(fs):
        mses[f] = ((data[idxs==f][response]
            -
            fitmodel(features,data[idxs!=f],response).predict(data[idxs==f]))**2).mean()
    return mses.mean(), mses.std(ddof=1)/np.sqrt(fs)
def PredError(model, df=df_test):
    return ((df['quality'] - model.predict(df))**2).mean()
def rsquaredCV(response, features, data):
    return 1-CV_MSE(response,features,data)[0] / data[response].var()
```

```python
## forward stepwise
def fwdSW(feat, data, response):
    Sfwd = set()
    features = set(feat)
    
    while len(Sfwd)<len(features):
        f = max(features - Sfwd, key = lambda f: mycriterion(fitmodel(Sfwd.union({f}),data,response)))
        after = mycriterion(fitmodel(Sfwd.union({f}),data,response))
        before = mycriterion(fitmodel(Sfwd,data,response))
        if after > before:
            Sfwd = Sfwd.union({f})
        else:
            break
    return Sfwd

def bckwdSW(feat, data, response):
    Sbwd = set(feat)
    
    while len(Sbwd)>0:
        f = max(Sbwd, key = lambda f: mycriterion(fitmodel(Sbwd-{f},data,response)))
        after = mycriterion(fitmodel(Sbwd-{f},data,response))
        before = mycriterion(fitmodel(Sbwd,data,response))
        if after > before:
            Sbwd = Sbwd-{f}
        else:
            break
    return Sbwd

def SWboth_calc(features, data, response):
    Sfwd = fwdSW(features, mycriterion, data, response)
    Sbwd = bckwdSW(features, mycriterion, data, response)
    Sboth = Sfwd if mycriterion(fitmodel(Sfwd,data,response)) > mycriterion(fitmodel(Sbwd,data,response)) else Sbwd
    
    return Sboth

def SWboth(Sfwd, Sbwd,data, response):
    Sboth = Sfwd if mycriterion(fitmodel(Sfwd,data,response)) > mycriterion(fitmodel(Sbwd,data,response)) else Sbwd 
    return Sboth

def getaic(lm):
    return -lm.aic
mycriterion = getaic
```
